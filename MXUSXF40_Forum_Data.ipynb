```python
import numpy as np
import matplotlib.pyplot as plt

# --- Constants and base parameters ---
D = 1.2           # Rotor diameter [m]
R = D / 2
V_base = 5.1      # Base windspeed [m/s]
TSR_base = 4      # Fixed TSR for second plot

# --- Range definitions ---
TSR_vals = np.linspace(3, 5, 200)               # TSR sweep for first plot
V_vals = np.linspace(V_base - 3, V_base + 3, 200)  # Windspeed sweep for second plot

# --- Calculations ---
# 1. RPM vs TSR (fixed windspeed)
omega_TSR = TSR_vals * V_base / R               # rad/s
rpm_TSR = omega_TSR * 60 / (2 * np.pi)

# 2. RPM vs Wind speed (fixed TSR)
omega_V = TSR_base * V_vals / R
rpm_V = omega_V * 60 / (2 * np.pi)

# --- Plot 1: TSR variation ---
plt.figure(figsize=(6,4))
plt.plot(TSR_vals, rpm_TSR, color='tab:blue')
plt.title("Rotor Speed vs TSR (V = 5.1 m/s)")
plt.xlabel("Tip-Speed Ratio (λ)")
plt.ylabel("Rotor Speed [RPM]")
plt.grid(True)
plt.tight_layout()

# --- Plot 2: Wind speed variation ---
plt.figure(figsize=(6,4))
plt.plot(V_vals, rpm_V, color='tab:orange')
plt.title("Rotor Speed vs Wind Speed (λ = 4)")
plt.xlabel("Wind Speed [m/s]")
plt.ylabel("Rotor Speed [RPM]")
plt.grid(True)
plt.tight_layout()

plt.show()
```
